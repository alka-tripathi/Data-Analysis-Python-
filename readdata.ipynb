```python
## now string back to json
df.to_json() ##orient bydefault take index
```

## Pickle file :
What is a pickle file in Python?

A pickle file is a file that stores Python objects in a binary format so they can be saved to disk and loaded back later exactly as they were.
This process is called serialization (save) and deserialization (load).



Why do we use pickle?

To save:

1) Pandas DataFrames

2) Lists, dictionaries, tuples

3) Trained ML models

4) Any Python object


```python
df_excel.to_pickle('df_excel')
```

```python
pd.read_pickle('df_excel')
```
