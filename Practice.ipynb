```python
df1=pd.DataFrame({'key1':[1,2,3,4],'value1':['A','B','C','D']})
df2=pd.DataFrame({'key2':[1,2,3,4],'value2':['A','B','C','D']})
```

```python
df1=pd.DataFrame({'key':[1,2,3,4],'value1':['A','B','C','D']})
df2=pd.DataFrame({'key':[1,2,3,4],'value2':['A','B','C','D']})
```

```python
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="left")
```
