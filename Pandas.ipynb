## Dataframe 
 as table with rows and column

```python
## Create dataframe with dictionary

## basically 3 column and 3 rows
data={
    'Name':['Alka','Tanay','Arisha'],
    'Age':[20,20,19],
    'City':['Bangalore','Dehli','New York']

}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

## Acessing the data from DataFram

üîπ When to Use What?

1)Use loc when working with labels or conditions

2)Use iloc when working with positions

df.loc[row_label, column_label]

df.iloc[row_index, column_index]


s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])

s.loc['b']   # 20
s.iloc[1]    # 20

## Accesing a specified element


at and iat are scalar accessors used to get or set a single value in a DataFrame or Series.

üëâ They are faster than loc and iloc because they work with only one cell.


loc ‚Üí multiple rows/columns, filtering

iloc ‚Üí slicing by position

at ‚Üí fastest single value (label-based)

iat ‚Üí fastest single value (position-based)

```python
df['Name']
```

```python
df
```

```python
print(df.at[1,'Name'])
```

```python
df.iat[2,2] ## row index and column index
```
